```python
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open('stats.json') as f:
    data = json.load(f)
```

```python
frames = [25000*i for i in range(1, 134)]
avg_score = []
max_score = []

for epoch in data:
    avg_score.append(epoch['expreplay/mean_score'])
    max_score.append(epoch['expreplay/max_score'])
```

```python
plt.figure(figsize=(20,10))
plt.plot(frames, avg_score, label = 'Average Score')
plt.plot(frames, max_score, label = 'Maximum Score')
plt.ylim([-25, 25])
plt.xlabel('Frames the Algorithm has trained on ~ Epochs')
plt.ylabel('Points (Max = 21)')
plt.title('Performance on a game of Pong')
plt.legend()
plt.savefig('pong_perfromance.jpeg')
```
